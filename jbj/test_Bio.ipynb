```python
from pylsl import StreamInlet, resolve_byprop

stream_names = ["PPG_GRN", "PPG_RED", "PPG_IR", 
                "EDA", 
                "ACC_X", "ACC_Y", "ACC_Z", 
                "GYRO_X", "GYRO_Y", "GYRO_Z", 
                "MAG_X", "MAG_Y", "MAG_Z", 
                "TEMP1", "THERM", "HR"]

# Inlet 딕셔너리 생성
inlets = {}
for name in stream_names:
    streams = resolve_byprop('name', name)
    if streams:
        inlets[name] = StreamInlet(streams[0])
        print(f"Connected to {name}")

# 데이터 수집
import time
while True:
    row = {}
    for name, inlet in inlets.items():
        sample, ts = inlet.pull_sample(timeout=0.0)  # non-blocking
        if sample is not None:
            row[name] = sample[0] if len(sample) == 1 else sample
    print(row)
    time.sleep(0.01)  # CPU 과부하 방지
```

```python
from pylsl import resolve_byprop

stream_names = ["PPG_GRN", "PPG_RED", "PPG_IR", 
                "EDA", 
                "ACC_X", "ACC_Y", "ACC_Z", 
                "GYRO_X", "GYRO_Y", "GYRO_Z", 
                "MAG_X", "MAG_Y", "MAG_Z", 
                "TEMP1", "THERM", "HR"]

for name in stream_names:
    streams = resolve_byprop('name', name)
    if streams:
        info = streams[0]
        print(f"{name}: {info.nominal_srate()} Hz")
```

```python
from pylsl import StreamInlet, resolve_byprop
import time, csv

stream_names = [
    "PPG_GRN", "PPG_RED", "PPG_IR", 
    "EDA", 
    "ACC_X", "ACC_Y", "ACC_Z", 
    "GYRO_X", "GYRO_Y", "GYRO_Z", 
    "MAG_X", "MAG_Y", "MAG_Z", 
    "TEMP1", "THERM", "HR"
]

# Inlet 딕셔너리 생성
inlets = {}
for name in stream_names:
    streams = resolve_byprop('name', name)
    if streams:
        inlets[name] = StreamInlet(streams[0])
        print(f"Connected to {name}")

# CSV 파일 준비
outfile = "emotibit_data.csv"
with open(outfile, "w", newline="") as f:
    writer = csv.writer(f)
    
    # 헤더
    header = ["Timestamp"] + list(inlets.keys())
    writer.writerow(header)
    
    print(f"Logging data to {outfile} ... Press Ctrl+C to stop.")
    
    try:
        while True:
            row = {}
            # 유닉스타임 (마이크로초까지, 소수점 6자리)
            timestamp = "{:.6f}".format(time.time())
            
            for name, inlet in inlets.items():
                sample, ts = inlet.pull_sample(timeout=0.0)
                if sample is not None:
                    row[name] = sample[0] if len(sample) == 1 else sample
                else:
                    row[name] = ""
            
            # 한 줄 기록
            writer.writerow([timestamp] + [row[n] for n in inlets.keys()])
            f.flush()
            
            time.sleep(0.01)
    
    except KeyboardInterrupt:
        print("\nStopped logging.")
```
