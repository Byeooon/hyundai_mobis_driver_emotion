```python
import pylsl
from pylsl import StreamInlet, resolve_streams

print("Looking for EmotiBit stream...")
streams = resolve_streams()  # 모든 LSL 스트림 탐색
print(streams)
```

```python

for s in streams:
    print(f"Name: {s.name()}, Type: {s.type()}, ID: {s.source_id()}")

# EmotiBit 스트림 선택 (예: 이름이 'EmotiBit' 또는 시리얼번호일 수 있음)
inlet = StreamInlet(streams[0])

print("Now receiving data...")
while True:
    sample, timestamp = inlet.pull_sample()
    print(timestamp, sample)
```

```python
from pylsl import StreamInlet, resolve_byprop

stream_names = ["PPG_GRN", "PPG_RED", "PPG_IR", 
                "EDA", 
                "ACC_X", "ACC_Y", "ACC_Z", 
                "GYRO_X", "GYRO_Y", "GYRO_Z", 
                "MAG_X", "MAG_Y", "MAG_Z", 
                "TEMP1", "THERM", "HR"]

# Inlet 딕셔너리 생성
inlets = {}
for name in stream_names:
    streams = resolve_byprop('name', name)
    if streams:
        inlets[name] = StreamInlet(streams[0])
        print(f"Connected to {name}")

# 데이터 수집
import time
while True:
    row = {}
    for name, inlet in inlets.items():
        sample, ts = inlet.pull_sample(timeout=0.0)  # non-blocking
        if sample is not None:
            row[name] = sample[0] if len(sample) == 1 else sample
    print(row)
    time.sleep(0.01)  # CPU 과부하 방지
```
